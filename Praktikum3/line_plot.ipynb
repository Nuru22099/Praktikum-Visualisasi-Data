```python
import matplotlib.pyplot as plt
```

```python
# Buat Data Sample
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Ags', 'Sep', 'Oct', 'Nov', 'Dec']
product_A_sales = [10,20,15,25,30,45,40,50,60,55,65,70]
product_B_sales = [5,10,8,15,18,20,22,30,25,35,40,45]
```

Line Plot Dengan Matplotlib

```python
# Line Plot 1 garis, style grafik default
plt.plot(months, product_A_sales, label='product A')
plt.title('Penjualan Produk A per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

Kustimasi Line Plot

```python
#kustomisasi line plot (Line 1 - Product A)
plt.plot(months, product_A_sales, label='Product A', linestyle="--", marker='o', color='blue')
#kustomisasi line plot (line 2 - Product B)
plt.plot(months, product_B_sales, label='Product B', linestyle="-", marker='x', color='red')
plt.title('Penjualan Produk per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

Multiple Line

```python
# Data Sample Tambahan
product_C_sales = [18,22,25,28,32,38,42,45,48,52,56,60]
product_D_sales = [7,9,11,13,16,18,20,23,25,28,30,33]

#plot Product A
plt.plot(months, product_A_sales, label='Product A', linestyle="--", marker='o', color='blue')

#plot Product B
plt.plot(months, product_B_sales, label='Product B', linestyle="-", marker='x', color='red')

#plot product C
plt.plot(months, product_C_sales, label='Product C', linestyle=":", color='green')
plt.title('Penjualan Produk per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

Garis Berbeda Untuk Menunjukan Trend

```python
# Line Plot
plt.plot(months, product_A_sales, label='Product A', linestyle="--", marker='o', color='blue')
plt.plot(months, product_B_sales, label='Product B', linestyle="-", marker='x', color='red')
plt.title('Penjualan Produk per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

Sublpot Untuk Menampilkan  Beberapa Grafik

```python
# Line Plot
fig, axs= plt.subplots(2, 1, figsize=(10, 8))

# Plot For Product A
axs[0].plot(months, product_A_sales, label='Product A', marker='o', color='blue')
axs[0].set_title('Penjualan Produk A per Bulan')
axs[0].set_xlabel('Bulan')
axs[0].set_ylabel('Jumlah Penjualan')
axs[0].legend()
axs[0].grid(True)

# Plot For Product B
axs[1].plot(months, product_B_sales, label='Product B', color='red', marker='x')
axs[1].set_title('Penjualan Produk B per Bulan')
axs[1].set_xlabel('Bulan')
axs[1].set_ylabel('Jumlah Penjualan')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
