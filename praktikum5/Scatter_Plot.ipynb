**Scater Plot**


**1. Import Library**



```python
import matplotlib.pyplot as plt
```

**2. Buat Dataset**

```python
suhu = [20, 22, 24, 26, 28, 30, 32, 34, 36]
penjualan =[50, 60, 70, 90, 100, 110, 130, 150, 180]
```

**3. Basic Scatter Plot**

```python
plt.scatter(suhu, penjualan)
plt.title('Hubungan Penjualana Es Krim Dengan Suhu')
plt.xlabel('Suhu')
plt.ylabel('Penjualan Es Krim')
plt.show()
```

**4. Kustomisasi Scatter Plot**

```python
plt.scatter(suhu, penjualan, color='red', marker='o', s=100, edgecolor='black', alpha=0.7)
plt.title('Hubungan Penjualana Es Krim Dengan Suhu')
plt.xlabel('Suhu')
plt.ylabel('Penjualan Es Krim')
plt.show()
```

**5. Multiple Scatter Plot**

```python
# Dataset Suhu
suhu = [20, 22, 24, 26, 28, 30, 32, 34, 36]

# Dataset Penjualan
penjualan_weekdays = [50, 60, 70, 80, 90, 100, 110, 120, 130]
penjualan_weekend = [60, 70, 80, 100, 110, 120, 140, 160, 200]

plt.scatter(suhu, penjualan_weekdays, color="pink", label="Hari Kerja", s=80)
plt.scatter(suhu, penjualan_weekend, color="blue", label="Hari Libur", s=80)
plt.title('Hubungan Penjualana Es Krim Dengan Suhu')
plt.xlabel('Suhu')
plt.ylabel('Penjualan Es Krim')
plt.legend()
plt.show()
```

**6. Melihat Hubungan Data dengan Variabel Lain**


```python
# Dataset
suhu = [20, 22, 24, 26, 28, 30, 32, 34, 36]
penjualan =[50, 60, 70, 90, 100, 110, 130, 150, 180]

# Dataset
kelembapan =[60, 65, 70, 75, 80, 85, 90, 95, 100]
ukuran_titik =[200, 220, 240, 260, 280, 300, 320, 340, 360]

# Kustomisasi Scatter Plot
plt.scatter(suhu, penjualan, c=kelembapan, s=ukuran_titik, cmap='coolwarm', alpha=0.6)
plt.title('Hubungan Penjualana Es Krim Dengan Suhu')
plt.xlabel('Suhu')
plt.ylabel('Penjualan Es Krim')
plt.colorbar(label='Kelembapan (%)')
plt.show()
```

**7. 3D Scatter Plot**

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Dataset Suhu
suhu = [20, 22, 24, 26, 28, 30, 32, 34, 36]

# Dataset Penjualan
penjualan_weekdays = [50, 60, 70, 80, 90, 100, 110, 120, 130]
penjualan_weekend = [60, 70, 80, 100, 110, 120, 140, 160, 200]

# 3D Scatter Plot
fig = plt.figure(figsize=(15, 13))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(suhu, penjualan_weekdays, penjualan_weekend,
           color='red', s=80, label='Data Penjualan')

ax.set_title('Hubungan Penjualan Es Krim Dan Suhu (3D ver)')
ax.set_xlabel('Suhu')
ax.set_ylabel('Penjualan Weekdays')
ax.set_zlabel('Penjualan Weekend')
ax.legend()
plt.show()
```
