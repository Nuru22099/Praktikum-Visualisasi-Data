**1. Import Library**

```python
import matplotlib.pyplot as plt
import numpy as np
```

**2. Dataset**

```python
# Data Penjualan
brands = ['Samsung', 'Iphone', 'Vivo', 'Oppo']
sales_2023 = [500, 700, 300, 400]
sales_2024 = [550, 750, 350, 450]
sales_2025 = [600, 800, 400, 500]
```

**3. Basic Chart (Horizontal & Stacked Horizontal Bar Chart)**

```python
y_pos = np.arange(len(brands))

# Canvas dan Axis Grafik
fig, ax1, = plt.subplots()

# Grafik Batang Horizontal
ax1.set_title('Horizontal Bar Chart - 2023')
ax1.set_xlabel('Jumlah Penjualan')
ax1.set_ylabel('Merk')
ax1.barh(brands, sales_2023)
plt.show()

# Staked Horizontal Bar Chart
# Canvas dan Axis Grafik
fig, ax2, = plt.subplots()

# Grafik Batang Horizontal
ax2.set_title('Staked Horizontal Bar Chart')
ax2.set_xlabel('Jumlah Penjualan')
ax2.set_ylabel('Merk')

# Bar Pertama Untuk 2023
ax2.barh(brands, sales_2023, label='2023')

# Bar Kedua Untuk 2025
ax2.barh(brands, sales_2023, left=sales_2023, label='2025')

ax2.legend()
plt.show()
```

**4. Kustomisasi Grafik**

```python
from matplotlib.lines import lineStyles
y_pos = np.arange(len(brands))
colors = ['gold', 'skyblue', 'coral', 'lightgreen']

# Custom Horizontal Bar Chart (2023)
fig, ax3 = plt.subplots()

ax3.barh(brands, sales_2023, color=colors, edgecolor='black', height=0.6)
ax3.set_title('Penjualan Smartphone Tahun 2023 (Custom)')
ax3.set_xlabel('Jumlah Penjualan')
ax3.set_ylabel('Merk')
ax3.grid(axis='x', linestyle='--', alpha=0.7)

# Menampilkan Label Nilai Di Bar Chart
for i, v in enumerate(sales_2023):
  ax3.text(v+10, i, str(v), color='black')

plt.show()

# Custom Staked Horizontal Bar Chart (2023-2024)
fig, ax4 = plt.subplots()

ax4.barh(brands, sales_2023, color='lightcoral', edgecolor='black', label='2023')
ax4.barh(brands, sales_2024, left=sales_2023, color='lightskyblue', edgecolor='black', label='2024')

ax4.set_title('Penjualan Smartphone Tahun 2023 & 2024 (Custom)')
ax4.set_xlabel('Jumlah Penjualan')
ax4.set_ylabel('Merk')
ax4.grid(axis='x', linestyle='--', alpha=0.5)

# Hitung Total Penjualan dari 2 Tahun Untuk Label Angka
total_sales = np.array(sales_2023) + np.array(sales_2024) + np.array(sales_2025)

# Label Nilai
for i, v in enumerate(total_sales):
  ax4.text(v+10, i, str(v), va='center')

plt.show()
```

**5. Multiple Chart**

```python
y_pos = np.arange(len(brands))

fig5, ax5 = plt.subplots()

# Bar Pertama
ax5.barh(y_pos - 0.2, sales_2023, color='pink', height=0.4, label='2023')

# Bar Kedua
ax5.barh(y_pos + 0.2, sales_2024, color='lightblue', height=0.4, label='2024')

ax5.set_yticks(y_pos, brands)
ax5.set_title('Penjualan Smartphone Tahun 2023 vs 2024 (Custom)')
ax5.set_xlabel('Jumlah Penjualan')
ax5.set_ylabel('Merk')
ax5.legend()
plt.show()

# Multiple Stacked Horizontal Bar Chart
fig6, ax6 = plt.subplots()
# Bar Pertama
ax6.barh(y_pos, sales_2023, color='skyblue', label='2023')
# Bar Kedua
ax6.barh(y_pos, sales_2024, left=sales_2023, color='salmon', label='2024')
# Bar Ketiga
ax6.barh(y_pos, sales_2025, left=np.array(sales_2023), color='lightgreen', label='2025')

ax6.set_yticks(y_pos, brands)
ax6.set_title('Penjualan Smartphone Tahun 2023 vs 2024 vs 2025 (Stacked)')
ax6.set_xlabel('Jumlah Penjualan')
ax6.set_ylabel('Merk')
ax6.legend()
plt.show()
```
